```python
#Imports
import plotly.graph_objects as go
import pandas as pd
```

```python
#Data processing
characters_df = pd.read_csv("./character-predictions_pose.csv")
print(characters_df)
characters_df['gender']=characters_df['male'].replace(1,'Male').replace(0,'Female')
characters_df=characters_df.replace(to_replace ='.*rotherhood.*', value = 'Brotherhood Without Banners', regex = True)
gender_df = characters_df.groupby(by=["gender","house"]).size().reset_index(name='counts')
pd.set_option('display.max_rows', gender_df.shape[0]+1)
```

```python
#Plot nodes selection
all_nodes = gender_df.gender.values.tolist() + gender_df.house.values.tolist()
source_indices = [all_nodes.index(Name) for Name in gender_df.gender]
target_indices = [all_nodes.index(House) for House in gender_df.house]
```

```python
#Plot away!
fig = go.Figure(data=[go.Sankey(
    # Define nodes
    node = dict(
      label =  all_nodes,
        pad = 20,
      thickness = 20,
    ),

    # Add links
    link = dict(
      source =  source_indices,
      target =  target_indices,
      value =  gender_df.counts,
))])

fig.update_layout(title_text="Gender distribution for different houses in the Game of Thrones",
                  font_size=10)
fig.show()
```
